# Visualization of expression data

When working with omics data, visualization of the expression dataset is an integral part of the workflow. The `ExpressionProfile` class offers a pretty nice interface for that exact purpose. We use [altair](https://github.com/altair-viz/altair) as our toolbox of choice for interactive visualization due to its simple and clean approach. Let's look at some examples:

```python
import numpy as np
from driven import ExpressionProfile

genes = ["Gene2", "Gene3", "Gene6", "Gene7"]
conditions = ["Exp#1", "Exp#2", "Exp#3"]
expression = np.array(([0.17, 0.20, 0.93],
                       [0.36, 0.83, 0.77],
                       [0.87, 0.65, 0.07],
                       [0.55, 0.49, 0.52]))

model_expression = ExpressionProfile(genes, conditions, expression)
```

## Histogram

```python
model_expression.hist()
```

## Scatter plot

```python
model_expression.scatter(x="Exp#1", y="Exp#2")
```

## Box plot

```python
model_expression.box()
```

## Heatmap

```python
model_expression.heatmap()
```
